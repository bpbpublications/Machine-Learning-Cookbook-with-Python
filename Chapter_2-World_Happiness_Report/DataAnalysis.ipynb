```python
def std_moments(data,n):
    data = data.values
    num = 0 
    std = 0
    mean = np.mean(data)
    for i in data:
        num += (i - mean)**n
        std += (i - mean)**2
#     return ((num)/(len(data)))/((np.sqrt(std/(len(data)+1)))**n)
    return ((num)/(len(data)))/((np.std(data))**n)
```

```python
import numpy as np
def box_plot_calculation(data):
    data = data.values
    q25, q75 = np.percentile(data, 25), np.percentile(data, 75)
    print('Quartile 25: {} | Quartile 75: {}'.format(q25, q75))
    IQR = q75 - q25
    print('IQR: {}'.format(IQR))

    cut_off = IQR * 1.5
    MIN, MAX = q25 - cut_off, q75 + cut_off
    print('Cut Off: {}'.format(cut_off))
    print('Minimum: {}'.format(MIN))
    print('Maximum: {}'.format(MAX))

    outliers = [x for x in data if x < MIN or x > MAX]
    outliers.sort()
    print('Feature Outliers: {}'.format(len(outliers)))
    print('Outliers:{}'.format(outliers))
```

```python
box_plot_calculation(hr["Dystopia Residual"])
```
